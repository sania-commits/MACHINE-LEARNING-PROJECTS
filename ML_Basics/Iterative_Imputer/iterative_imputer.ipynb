# Through Looping:

```python
df= np.round(pd.read_csv('50_Startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
np.random.seed(9)
df = df.sample(5)
df
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imp = IterativeImputer(random_state=0)

df_imputed = imp.fit_transform(df)
```

```python
df_imputed
```

```python
df.isnull().sum()
```

```python
df.loc[df.index[0], 'R&D Spend'] = np.nan
df.loc[df.index[2], 'Administration'] = np.nan
df.loc[df.index[4], 'Marketing Spend'] = np.nan
```

```python
df
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imp = IterativeImputer(random_state=0)

df_imputed = pd.DataFrame(
    imp.fit_transform(df),
    columns=df.columns,
    index=df.index
)
```

```python
print("Original DataFrame")
print(df)

print("\nImputed DataFrame")
print(df_imputed)
```

```python
df_original = df.copy()
```

```python
print(df_original)
print(df_imputed)
```

```python
for i in [1, 2, 5, 10]:
    imp = IterativeImputer(max_iter=i, random_state=0)

    df_imp = pd.DataFrame(
        imp.fit_transform(df),
        columns=df.columns
    )

    print(f"\nmax_iter = {i}")
    print(df_imp)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

imp = IterativeImputer(
    estimator=RandomForestRegressor(
        n_estimators=100,
        random_state=0
    ),
    random_state=0
)

df_rf = pd.DataFrame(
    imp.fit_transform(df),
    columns=df.columns,
    index=df.index
)

print(df_rf)
```

```python
imp1 = IterativeImputer(random_state=0)

df_br = pd.DataFrame(
    imp1.fit_transform(df),
    columns=df.columns,
    index=df.index
)
```

```python
imp2 = IterativeImputer(
    estimator=RandomForestRegressor(random_state=0),
    random_state=0
)

df_rf = pd.DataFrame(
    imp2.fit_transform(df),
    columns=df.columns,
    index=df.index
)
```

```python
print("Bayesian Ridge")
print(df_br)

print("\nRandom Forest")
print(df_rf)
```
