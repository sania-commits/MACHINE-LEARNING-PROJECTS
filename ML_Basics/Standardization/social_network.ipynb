```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
# Train test split:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1), df['Purchased'], test_size=0.2, random_state=0)
X_train.shape, X_test.shape
```

```python
#StandardScaler:
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# fit the sc to the train set, it will learn the parameters
sc.fit(X_train)
# transform train and test sets
X_train_scaled = sc.transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
# Effect of Scaling:
fig,(ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))
ax1.scatter(X_train['Age'],X_train['EstimatedSalary'])
ax2.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'])
ax1.set_title('Before Scaling')
ax2.set_title('After Scaling')
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))
# Before scaling:
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1, color='blue', label='Age')
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1, color='red', label='EstimatedSalary')
ax1.legend()    
# After scaling:
ax2.set_title('After Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2, color='blue', label='Age')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2, color='red', label='EstimatedSalary')
ax2.legend()
plt.show()
```

```python
# Comparison of distributions:
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))
# Before scaling:
ax1.set_title('Age Distribution BeforeScaling')
sns.kdeplot(X_train['Age'], ax=ax1, color='blue', label='Age')
ax1.legend()
#After scaling:
ax2.set_title('Age Distribution After Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2, color='orange', label='Age')
ax2.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Salary Distribution Before Scaling')
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('Salary Distribution Standard Scaling')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

```python
# Why Scaling is Important?
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print('Accuracy before scaling:', accuracy_score(y_test, y_pred)*100)
print('Accuracy after scaling:', accuracy_score(y_test, y_pred_scaled)*100)
```

```python
print('confusion_matrix before scaling:\n', confusion_matrix(y_test, y_pred))
print('confusion_matrix after scaling:\n', confusion_matrix(y_test, y_pred_scaled))
```

```python
# Effect of Outliers:
df = pd.concat([df,pd.DataFrame({'Age':[5,90,95],'EstimatedSalary':[1000,25000,35000],'Purchased':[0,1,1]})],ignore_index=True)
```

```python
df
```

```python
plt.scatter(df['Age'],df['EstimatedSalary'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.3,random_state=0)
X_train.shape,X_test.shape
```

```python
# Standard Scaler:
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))
ax1.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'])
ax1.set_title('Before Scaling with Outliers')
ax2.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'])
ax2.set_title('After Scaling with Outliers')
plt.show()
```
