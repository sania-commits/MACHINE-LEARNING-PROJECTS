# Working with Dates:

```python
date['date'] = pd.to_datetime(date['date'])
time['date'] = pd.to_datetime(time['date'])
```

```python
date.info()
```

```python
time.info()
```

# 1. Extract Year:

```python
date['date_year'] = date['date'].dt.year
date.sample(5)
```

# 2. Extract Month_number:

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

# 3. Extract Month_name:

```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

# 4. Extract Day:

```python
date['date_day'] = date['date'].dt.day
date.head()
```

# 5. Extract Day of Week:

```python
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

# 6. Extract Day of week name:

```python
date['date_dow_name'] = date['date'].dt.day_name()
date.drop(columns=['product_id','city_id','orders'])
date.head()
```

# 7. Checking if it's Weekend?

```python
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','orders'])
date.head()
```

# 8. Extract week of the year:

```python
date['date_week']= date['date'].dt.isocalendar().week
date.head()
```

# 9. Extract Quarter:

```python
date['quarter'] = date['date'].dt.quarter
date.head()
```

# 10. Extract Semester:

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
date.head()
```

# Extract Time elapsed between Dates:

```python
import datetime
```

```python
today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months Passed:

# today is a Python datetime.datetime object, and to_period() is a pandas method, not a Python datetime method.
# Convert today to a pandas Timestamp
date['months_passed']= (pd.Timestamp(today).to_period('M') - date['date'].dt.to_period('M')).apply(lambda x: x.n)
date.head()
```
